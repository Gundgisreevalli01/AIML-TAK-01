```python
import pandas as pd
import numpy as np

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# 1. Display first and last records
print("First 5 records:")
print(df.head())
print("\nLast 5 records:")
print(df.tail())

# 2. Manual Identification (Commented logic)
# Numerical: Age, Fare, SibSp, Parch
# Categorical: Embarked
# Ordinal: Pclass (1st, 2nd, 3rd class)
# Binary: Survived, Sex

# 3. Data types, null values, and statistical summaries
print("\n--- Dataset Info ---")
df.info()

print("\n--- Numerical Summary ---")
print(df.describe())

print("\n--- Categorical Summary ---")
print(df.describe(include=['O']))

# 4. Unique values in categorical columns
print("\n--- Unique Values ---")
cat_cols = ['Survived', 'Pclass', 'Sex', 'Embarked']
for col in cat_cols:
    print(f"{col}: {df[col].unique()}")

# 5. ML Suitability
target_variable = 'Survived'
input_features = df.drop(columns=['Survived', 'PassengerId', 'Name', 'Ticket']).columns.tolist()

print(f"\nTarget Variable: {target_variable}")
print(f"Input Features: {input_features}")
```
